```python
import pandas as pd
import numpy as np
```

```python
#read result4.csv
df = pd.read_csv('result3.csv')

#calculate 1st Author Paper Count/ 1st Author Citation Count
df['publication_citation rate'] = df['1st Author Paper Count']/df['1st Author Citation Count']
```

```python
df.head(10)
```

```python
#mean of publication_citation rate(drop NaN and Inf)
k = df['publication_citation rate'].replace([np.inf, -np.inf], np.nan)
k = k.dropna()
m = k.mean()
v = k.var()
print(m, v)
```

```python
#add a column calle '1st Author Ordinal'
df['1st Author Ordinal'] = 0

#set the value of '1st Author Ordinal' column, if the value of 'publication_citation rate' is larger than the mean + variance, then set the value of '1st Author Ordinal' to "one star"
#for each row in df
for index, row in df.iterrows():
    if (row['publication_citation rate'] >= m + v) & (row['publication_citation rate'] < 1):
        df.at[index, '1st Author Ordinal'] = "one star"
    elif (row['publication_citation rate'] <= m - v) & (row['publication_citation rate'] > 0):
        df.at[index, '1st Author Ordinal'] = "three star"
    else:
        df.at[index, '1st Author Ordinal'] = "two star"
```

```python
#save it to result4.csv
df.to_csv('result4.csv', index=False)
```
